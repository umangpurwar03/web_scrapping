#WEB SCRAPPING

Import some of basic libraries of web scrapping like beautifulsoup and need to clean the data using NLTK library

```python
import numpy as np
import pandas as pd
import requests
import bs4 as bf
from bs4 import BeautifulSoup as bfs
import nltk
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
url='https://platform.fynd.com/blog/business-ideas-in-patna'
```

```python
link=requests.get(url)
link
```

```python
html=bfs(link.content,"lxml")
html
```

```python
heading = html.find(attrs={"class":"blog-top-content"}).text
heading
```

```python
print(len(html))
try:
 articles=(html.find(attrs={"class":"body"}).text)
except Exception as err:
    print(f"error was {err}")
articles
```

```python
articles= articles.replace('\n','')
articles
```

```python
articles= articles.replace('\t','')
articles
```

```python
nltk.download('punkt')
```

```python
nltk.download('stopwords')
```

```python
stop_words = list(set(stopwords.words('english')))
```

```python
sentence=len(sent_tokenize(articles))
sentence
```

```python
cleaned_articles = [' ']*len(articles)
```

```python
cleaned_articles= articles.replace(' '+w+' ',' ').replace('?','').replace('.','').replace(',','').replace('!','')
cleaned_articles
```
